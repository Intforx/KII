```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os

# 指定你的 PDF 所在文件夹（从 notebook 往上跳一层，再进去子文件夹）
pdf_folder = "../Quelle"
all_documents = []

# 遍历加载 PDF
for filename in os.listdir(pdf_folder):
    if filename.endswith(".pdf"):
        loader = PyPDFLoader(os.path.join(pdf_folder, filename))
        docs = loader.load()
        all_documents.extend(docs)

print(f"📄 共加载了 {len(all_documents)} 页艺术史内容！")
```

```python
# ✅ Step 2: 文本分段（chunking）
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_docs = text_splitter.split_documents(all_documents)
print(f"✂️ 共分成 {len(split_docs)} 个段落")

# ✅ Step 3: 提取纯文本
texts = [doc.page_content for doc in split_docs]

# ✅ Step 4: 加载 BGE-M3 模型（适用于 flagembedding==1.0.7）
model = FlagModel("BAAI/bge-m3")

# ✅ Step 5: 生成前100段向量（可改为 texts 全部）
vectors = model.encode(texts[:100])  # 改成 texts 就是全量
```

```python
from langchain.vectorstores import FAISS
from langchain.docstore.document import Document

# 创建一个对应的 Document 列表（绑定文本和元信息）
docs_for_faiss = [Document(page_content=texts[i], metadata={"source": f"chunk_{i}"}) for i in range(len(vectors))]

# 创建 FAISS 向量库
faiss_db = FAISS.from_embeddings(vectors, docs_for_faiss)

# 保存数据库到本地（方便以后用）
faiss_db.save_local("kunstgeschichte_faiss_index")

print("✅ 向量已成功保存到 FAISS 向量数据库！")
```

```python
from langchain.vectorstores import FAISS

# ✅ 加载模型（你之前已经下载过 BGE-M3 了）
from FlagEmbedding import FlagModel
model = FlagModel("BAAI/bge-m3")

# ✅ 创建一个“假embedding类”，用于加载FAISS并计算查询向量
class DummyEmbedding:
    def embed_query(self, text):
        return model.encode([text])[0]  # 单句编码返回向量

# ✅ 加载本地FAISS数据库
faiss_db = FAISS.load_local("kunstgeschichte_faiss_index", embeddings=DummyEmbedding())

# ✅ 输入你想问的问题（可中/英/德！）
query = "Was sind die Merkmale der Barockkunst?"  # 可以换成：什么是文艺复兴？

# ✅ 从数据库中找出最相关的段落
results = faiss_db.similarity_search(query, k=3)

# ✅ 显示检索到的内容（片段原文）
for i, res in enumerate(results):
    print(f"\n📚 匹配结果 {i+1}：\n{res.page_content}\n")
```

```python
from langchain.vectorstores import FAISS
from FlagEmbedding import FlagModel

# ✅ 加载模型
model = FlagModel("BAAI/bge-m3")

# ✅ 修复后的 DummyEmbedding（支持函数调用）
class DummyEmbedding:
    def embed_query(self, text):
        return model.encode([text])[0]
    
    def __call__(self, text):
        return self.embed_query(text)

# ✅ 加载 FAISS 数据库，允许加载 .pkl
faiss_db = FAISS.load_local(
    "kunstgeschichte_faiss_index",
    embeddings=DummyEmbedding(),
    allow_dangerous_deserialization=True
)

# ✅ 输入问题
query = "Was sind die Merkmale der Barockkunst?"

# ✅ 检索相关段落
results = faiss_db.similarity_search(query, k=3)

# ✅ 显示结果
for i, res in enumerate(results):
    print(f"\n📚 匹配结果 {i+1}：\n{res.page_content}\n")
```
